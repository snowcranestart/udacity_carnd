###### gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
- Equalize the gray scale image to further reduce the light affection
###### equalized_img = cv2.equalizeHist(gray_img)
- Normalize the image to better converge during training
###### min_p, max_p = np.min(equalized_img), np.max(equalized_img)
###### normalized_img = ((equalized_img - min_p)/(max_p - min_p))*2 - 1
